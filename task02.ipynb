```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("tested.csv")
df['Age'].fillna(df['Age'].median(), inplace=True)
df['Fare'].fillna(df['Fare'].median(), inplace=True)
df.drop(columns='Cabin', inplace=True)

sns.set(style="whitegrid")
fig, ax = plt.subplots(1, 3, figsize=(18, 5))
sns.countplot(data=df, x='Survived', ax=ax[0])
ax[0].set_title("Overall Survival Count")
ax[0].set_xticklabels(['Did Not Survive', 'Survived'])

sns.countplot(data=df, x='Sex', hue='Survived', ax=ax[1])
ax[1].set_title("Survival by Sex")

sns.countplot(data=df, x='Pclass', hue='Survived', ax=ax[2])
ax[2].set_title("Survival by Passenger Class")

plt.tight_layout()
plt.show()
fig, ax = plt.subplots(2, 2, figsize=(16, 12))
sns.histplot(data=df, x='Age', hue='Survived', multiple='stack', kde=True, ax=ax[0, 0])
ax[0, 0].set_title("Survival Distribution by Age")
sns.boxplot(data=df, x='Pclass', y='Fare', ax=ax[0, 1])
ax[0, 1].set_title("Fare Distribution by Passenger Class")
sns.countplot(data=df, x='Embarked', hue='Survived', ax=ax[1, 0])
ax[1, 0].set_title("Survival by Embarkation Port")
numeric_df = df.select_dtypes(include=['int64', 'float64'])
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm', ax=ax[1, 1])
ax[1, 1].set_title("Correlation Heatmap")

plt.tight_layout()
plt.show()
```
