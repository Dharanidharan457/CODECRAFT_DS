```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset (adjust the path if needed)
df = pd.read_csv("population_data.csv", skiprows=4)

# Extract the 2023 population data and clean
pop_2023 = df[['Country Name', '2023']].dropna()
pop_2023.columns = ['Country', 'Population_2023']

# Plot histogram
plt.figure(figsize=(10, 6))
plt.hist(pop_2023['Population_2023'], bins=30, color='skyblue', edgecolor='black')
plt.title('Distribution of Countries by Population Size (2023)')
plt.xlabel('Population')
plt.ylabel('Number of Countries')
plt.grid(True)
plt.tight_layout()
plt.show()
```
